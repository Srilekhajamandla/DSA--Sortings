### Bubble sort 
- recursively swapping adjecent values 

#### Steps
- lowest
- swapping
- inc/dec range

```python

arr=[4,2,1,5]
def bubble_sort():
    n=len(arr)
    for i in  range(n+1): # to loop through count of list 
        for j in range(0,n-i-1): # to check all adjacent elements
        # i is the count
        # j is the element
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]#swap a,b=b,a
                
    
    return arr
```

## Selection Sort

- goal:=   finding lowest one
- store 1st element in a variable
- then check stored element with all the other elements
- if i is lowest then swap
- check index or values you are checking with

```python
arr=[5,4,3,2,1] 
n=len(arr)
for i in range(n):
    lowest=i
    
    for j in range(i+1,n):
        
        if arr[j]<arr[lowest]:
            
            lowest=arr[j]
    arr[i],arr[lowest]=arr[lowest],arr[i]
    
print(arr)    
```

## Binary search

- 1st finding low , high, mid => (low+high)/2
- if arr[i] < mid => update low, high, mid (check leftside)
- if arr[i]> mid => update low, high, mid (check rightside)
- check low<= high
- we dont how many no.of iterations , son use while loop
-  low=high+1, which breaks the condition

```python
arr = [2,3,5,7,8,10,12]

target = 7

low = 0
high = len(arr) - 1

while low <= high:

    mid = (low + high) // 2

    if arr[mid] == target:
        print("Element found at index:", mid)
        break

    elif arr[mid] < target:
        low = mid + 1

    else:
        high = mid - 1

else:
    print("Element not found")
```

## Insertion Sort

- 1st we need to assume  1st element is sorted
- check the current and previous one if for range(1,n):
- shift element to right which is bigger
- insert curent element in correct position

START

FOR i = 1 to n-1

    x= arr[i]
- store the current element
    j = i - 1
- to check the previous element with the current element so we set j to i-1

    WHILE j >= 0 AND arr[j] > x
- j index should be > 0 otherwise it will be go like out of index range i.e(-1) and if j=-1 that means that space is empty , there is no element in that space 

        arr[j + 1] = arr[j]
- shift the current element to right position , that means arr[j] current element value stores in arr[j+1]=> right position
        j = j - 1
- j is setting to be empty space 

    END WHILE
- after the end of  the loop ,insert x value[the element we r checking with all the elements] into the postion of arr[j+1] i.e a empty space after shifting that element to right
    arr[j + 1] = x

END FOR

STOP

```python
arr = [5, 4, 3, 2, 1]

n = len(arr)

for i in range(1, n):

    key = arr[i] # current element
    j = i - 1 # previous element

    while j >= 0 and arr[j] > key: # checking arr[j]=>elements in arr > key => 1st element(changes for every i)

        arr[j + 1] = arr[j] # shifting to the right position if the values is greater 4, 5 = 5 ,5
        j -= 1   # reseting the shifted element place  -1,5

    arr[j + 1] = key  #-1+1=0 i.e 1st element or previous element

print(arr)
```
